# Dataset

```python
from tqdm import tqdm

def get_indx(y_range=(1979, 2020), delay=6, pen_x=6, last_p=6):
  n, res = delay+pen_x, []
  for y in range(y_range[0], y_range[1]):
    for m in range(1, 13):
      for p in range(1, 7):
        time = (y-1979)*72 + (m-1)*6 + p-1 - n
        if time >= 0:
          x_t = []
          for i in range(time, time+pen_x):
            year, month, decade = i//72+1979, i%72//6+1, i%6+1
            x_t.append((year, month, decade))
          res.append((x_t, (y, m, p)))
          if y==y_range[1]-1 and m == 12 and p > last_p:
            res.pop()
  return res

def build_ds(DF, inds, target=(43.0, 132.0)):
  trg = DF[(DF.Latitude == target[0]) & (DF.Longitude == target[1])]
  # dt = DF[(DF.Latitude != target[0]) | (DF.Longitude != target[1])]
  dt = DF
  res = pd.DataFrame()
  for i in tqdm(inds):
    y = trg.loc[(trg.Year == i[1][0]) & (trg.Month == i[1][1]) & (trg.Record == i[1][2])]
    X, t = [], True
    for j in i[0]:
      X.append(dt.loc[(dt.Year == j[0]) & (dt.Month == j[1]) & (dt.Record == j[2])])
      if X[-1].empty: t = False
    if t and not y.empty:
      row = {'y': float(y.Value_ta)}
      for r, x in enumerate(X):
        for p, j in enumerate(x.Value_u):
          row['x_U_'+str(r)+'_'+str(p)] = j
        for p, j in enumerate(x.Value_v):
          row['x_V_'+str(r)+'_'+str(p)] = j
        for p, j in enumerate(x.Value_ta):
          row['x_Ta_'+str(r)+'_'+str(p)] = j
        for p, j in enumerate(x.Value_tr):
          row['x_Tr_'+str(r)+'_'+str(p)] = j
        for p, j in enumerate(x.Value_prs):
          row['x_Prs_'+str(r)+'_'+str(p)] = j
        for p, j in enumerate(x.Value_ice):
          row['x_Ice_'+str(r)+'_'+str(p)] = j
        for p, j in enumerate(x.Value_stp):
          row['x_Stp_'+str(r)+'_'+str(p)] = j
      row = pd.DataFrame([row])
      res = pd.concat([res, row], ignore_index=True)
  return res
```

```python
data = pd.read_csv('data_2.csv', parse_dates=['DateTime'])
data = data[(data.Latitude >= 42.5) & (data.Latitude <= 43.5)]
data = data[(data.Longitude >= 131.0) & (data.Longitude <= 133.5)]
data.describe()
```

```python
data['Year'] = data.DateTime.dt.year
data['Month'] = data.DateTime.dt.month
data['Record'] = data.DateTime.dt.day
data.Record[data.Record==6] = 2
data.Record[data.Record==11] = 3
data.Record[data.Record==16] = 4
data.Record[data.Record==21] = 5
data.Record[data.Record==26] = 6
data.describe()
```

```python
inds = get_indx(last_p=4)
dataset = build_ds(data, inds)
dataset
```
